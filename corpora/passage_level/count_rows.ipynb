```python
def count_jsonl_items(file_path):
    """统计JSONL文件的item数量（过滤空行）"""
    count = 0
    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            # 过滤空行（去除前后空格后仍不为空才计数）
            if line.strip():
                count += 1
    return count

# 使用示例
file_path = "cloud.jsonl"  # 替换为你的JSONL文件路径
total_items = count_jsonl_items(file_path)
print(f"JSONL文件共有 {total_items} 个item")
```
